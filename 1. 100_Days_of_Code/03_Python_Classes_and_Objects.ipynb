````python
class Tag:
    def __init__(self,text,tag):
        self.text=text
        self.tag=tag
    def render(self):
        return f"<{self.tag}>{self.text}</{self.tag}>"
````

````python
class Paragraph(Tag):#this defines inheritance from parent
    def __init__(self,text):
        Tag.__init__(self,text,tag='p')
````

````python
class ListItem(Tag): #this defines inheritance from parent
    def __init__(self,text):
        Tag.__init__(self,text,tag='li')
````

````python
a="I am a paragrpah."
````

````python
b=Paragraph(text=a)
````

````python
b.render()
````

````python
c='I am a list of item'
````

````python
d=ListItem(text=c)
````

````python
d.render()
````

````python
from abc import ABC, abstractclassmethod
#abstract classes and polymorphism, which are core concepts in object-oriented design

class Shape(ABC):
    @abstractclassmethod
    
    def draw(self):
        pass


class Eclipse(Shape):
    def draw(self):
        return 'Drawing an Eclipse'
class Rectangle(Shape):
    def draw(self):
        return 'Drawing a Rectangle'
````

````python
shapes=[Eclipse(),Rectangle()]
````

````python
for shape in shapes:
    print(shape.draw())
````

````python
r=Rectangle(5,3)
````

````python
class Rectangle(Shape):
    def __init__(self,width,length,char='X'):
        self.width=width
        self.length=length
        self.char=char
    
    def __str__(self):
        lines = [self.char * self.width for _ in range(self.length)]
        return "\n".join(lines)

    def draw(self):
        for i in range(self.length):
            print(self.char*self.width)
    def __repr__(self):
        return f"Rectangle(width={self.width}, length={self.length}, char='{self.char}')"
````
