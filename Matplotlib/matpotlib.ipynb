#### Title Lables Limits

```python
#Matlab Style
plt.style.use('seaborn')

plt.plot(month_number,interest_paid,c='k',label='Interest')
plt.plot(month_number,principal_paid,c='b',label='Principal')
plt.xlim(left=1,right=70)
plt.ylim(bottom=0,top=700)
plt.xlabel('Month',fontsize=18)
plt.ylabel('Dollars',fontsize=18)
plt.title('Interest and Principal Paid',fontsize=25)
plt.grid(b=True)
#plt.facecolor('w')
#axes.grid(axis='x',c='g',alpha=0.9,linestyle='-')

#plt.legend()
#plt.legend(loc='center right')
plt.legend(loc=[1.02,0])
plt.tight_layout()
plt.savefig('plot_matplotlib.png')
```

```python
#Object Oriented
fig,axes=plt.subplots(nrows=1,ncols=1)
axes.plot(month_number,interest_paid,c='k',label='Interest')
axes.plot(month_number,principal_paid,c='b',label='Principal')
axes.set_xlim(left=1,right=70)
axes.set_ylim(bottom=0,top=700)
axes.set_xlabel('Month',fontsize=18)
axes.set_ylabel('Dollars',fontsize=18)
axes.set_title('Interest and Principal Paid',fontsize=25)
axes.grid(axis='x')
axes.set_facecolor('w')
#axes.legend()
axes.legend(loc='center right')
#axes.legend(loc=[1.02,0])
plt.tight_layout() #to not get cutoff legends and Months
fig.savefig('plot_matplotlib1.png')
```
