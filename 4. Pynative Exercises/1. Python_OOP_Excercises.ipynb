```python
import math
class Shape:
    def area(self):
        raise NotImplementedError("Area method must be implemented by subclasses")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        area1=math.pi*self.radius*self.radius
        return f"Area of the Circle is {area1}"
    

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        area2=self.side*self.side
        return f"Area of the Square is {area2}"

# Example of polymorphism
shapes = [Circle(5), Square(7), Circle(3)]

for shape in shapes:
    print(shape.area())  # Output: 78.53975, 49, 28.27431
```
