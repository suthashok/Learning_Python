#### Fixed Code

```python
import random
guess = ''
while guess not in ('heads', 'tails'):
    print('Guess the coin toss! Enter heads or tails:')
    guess = input().lower()  #Ensured User Input to be lowercase for comparision purpose

toss = random.randint(0, 1)  # 0 is tails, 1 is heads

# Changed toss variable from 0/1 to heads/tails
if toss==0:
    toss='tails'
else:
    toss='heads'

if toss == guess:
    print('You got it!')
else:
    print('Nope! Guess again!')
    guess = input().lower() #Ensured User Input to be lowercase for comparision purpose
    if toss == guess:
        print('You got it!')
    else:
        print('Nope. You are really bad at this game.')
```
