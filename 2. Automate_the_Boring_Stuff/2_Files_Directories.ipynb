```python
with open('mad_libs.txt','w') as file:
    file.write('The ADJECTIVE panda walked to the NOUN and then VERB. A nearby NOUN was unaffected by these events.')
```

```python
def mad_libs(file_name):
    with open(file_name) as file:
        content=file.read()
    adj=input('Enter an adjective:\n')
    noun=input('Enter a noun:\n')
    verb=input('Enter a verb:\n')
    noun2=input('Enter a noun:\n')
    content=content.replace('ADJECTIVE',adj)
    content=content.replace('NOUN',noun,1)
    content=content.replace('VERB',verb)
    content=content.replace('NOUN',noun2)
    print(content)
    with open('mad_libs_op.txt','w') as file:
        file.write(content)
```

```python
mad_libs('mad_libs.txt')
```

#### Regex Search

```python
def regex_search(file_path,exp):
    p=Path(file_path)
    items=list(p.glob('*.txt'))
    for item in items:
        with open(item) as file:
            content=file.read()
            pattern = re.compile(rf'{exp}')
            match=pattern.search(content)
            print(match.group())
```

```python
regex_search(Path.cwd(),'\w')
```
