<h6>
Define a function without regular Expressions which determines whether given input is US phone number or not?
</h6>

```python
def is_phone_number(text):
    ''' US Phone Numbers are 10 Digits and Two Hypens. AAA-BBB-CCCC'''
    if len(text)==12 and '-' in text:
        c2=len([i for i in text if i=='-']) #Count of '-'
        a,b,c=text.split('-')
        c3=len(a)
        c4=len(b)
        c5=len(c)
        return c2==2 and c3==3 and c4==3 and c5==4
    else:
        return False
```

```python
is_phone_number('415-555-4242')
```

```python
is_phone_number('Moshi moshi')
```

<h6>
Define a function using regular Expressions which determines whether given input is US phone number or not?
</h6>

```python
def is_phone_num_re(text):
    ph_num=re.compile(r'\d{3}-\d{3}-\d{4}')
    if ph_num.search(text)!=None:
        return True
    return False
```

<h6>
    Strong Password Detection
</h6>

```python
def strong_password(text):
    if len(text)==8:
        return re.search(r'[a-z]',text)!=None and re.search(r'[A-Z]',text)!=None and re.search(r'[1-9]',text)!=None
    else:
        return False
```

```python
strong_password('asfusibd8')
```

```python
strong_password('asAusib8')
```

```python
strong_password('a@Ausib8')
```

<h6>
    Regrex Strip
</h6>

```python
def regex_strip(text,chars=None):
    ws_pattern=re.compile(r'^\s+|\s+$')
    char_pattern=re.compile(rf'[{chars}]+')
    if chars==None:
        return ws_pattern.sub('',text)
    else:
        return char_pattern.sub('',text)
```

```python
regex_strip('     Hello World!')
```

```python
regex_strip('     Hello World!','Wo')
```
